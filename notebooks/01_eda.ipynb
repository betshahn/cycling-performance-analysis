```python
# Cycling Performance Analysis (EDA)

## Goal
Understand what factors influence cycling speed using Strava activity data.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/betsyhahn/Downloads/cycling-performance-analysis/data/raw/public_strava_dataset.csv')
df.head()
```

```python
df_bike = df[df["type"] == "Ride"].copy()
df_bike.shape
```

```python
df_bike.info()
df_bike.describe()
```

```python
df_bike["average_speed"].describe()

## Target Variable: Average Speed
We use average speed as a proxy for cycling performance.
```

```python
## Speed distribution
plt.hist(df_bike["average_speed"], bins=15)
plt.title("Distribution of Cycling Speed")
plt.xlabel("Speed")
plt.ylabel("Count")

plt.savefig("figures/speed_distribution.png", bbox_inches="tight")
plt.show()
```

```python
## Elevation vs speed
plt.scatter(
    df_bike["total_elevation_gain"],
    df_bike["average_speed"]
)

plt.title("Speed vs Elevation Gain")
plt.xlabel("Elevation Gain")
plt.ylabel("Average Speed")

plt.savefig("figures/elevation_vs_speed.png", bbox_inches="tight")
plt.show()
```

```python
## Heart rate vs speed
plt.scatter(df_bike["average_heartrate"], df_bike["average_speed"])
plt.title("Speed vs Heart Rate")
plt.xlabel("Average Heart Rate")
plt.ylabel("Average Speed")

plt.savefig("figures/heart_rate_vs_speed.png", bbox_inches="tight")
plt.show()
```

```python
cols = [
    "average_speed",
    "average_heartrate",
    "total_elevation_gain",
    "distance",
    "moving_time"
]

df_bike[cols].corr()
```

```python
## What's been found (so far)

- Average heart rate is moderately positively correlated with speed
- Elevation gain is negatively correlated with speed
- Distance is weakly related to speed
- Performance is driven more by effort and terrain than ride length
```
